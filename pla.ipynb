```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

This is my target function

```python
def target(x):
    return 1-x
```

In this part I will generate 2-d dataset of random numbers ranging between -100 and 100. Then for loop labels the points randomly. Then we can visualize the data and target function, using pyplot library.

```python
n = 20 #change according to the exercise
#generating random numbers in 2-d
x1 = np.zeros(n)
x2 = np.zeros(n)
y = np.zeros(n)

for i in range(n):
    x1[i] = random.randint(-100, 100)
    x2[i] = random.randint(-100, 100)
    if x2[i]>target(x1[i]):
        y[i] = 1
    else:
        y[i] = -1
        
a = np.arange(-100,100)
plt.plot(a, target(a))
for i in range(0, n):
    if y[i] == 1:
        plt.scatter(x1[i],x2[i], color = 'blue', marker = '+')
    else:
        plt.scatter(x1[i],x2[i], color = 'red', marker = '$-$')
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")
plt.title("Data set size = %s" %n)
plt.show()
```

In this part I generated X matrix, added column of ones. Then generated Y matrix out of y_labels.

```python
X = np.zeros((n, 3))  #generate X matrix

for i in range(0, n):
    for j in range(0, 3):
        if j == 0:
            X[i][j] = 1
        elif j == 1:
            X[i][j] = x1[i]
        elif j == 2:
            X[i][j] = x2[i]
#print(X)

Y = np.zeros(n)
for k in range(0, n):
    Y[k] = y[k]
#print(Y)
```

This is the perceptron learning algorithm. Inputs are training dataset and labels. w = vector of weights. n_error = to keep track of the errors, correct_pred = to keep track of the correct predictions. 
The simple weight update equation is implemented. 

```python
def PLA(X, Y):
    w = np.zeros(3)
    correct_pred = 1  
    while correct_pred > 0:
        correct_pred = 1
        for i in range(0, len(X)):
            if (w[0] + w[1]*X[i][1] + w[2]*X[i][2])*Y[i] <= 0:
                  w = w + X[i]*Y[i]
                  break
            else:
                correct_pred = correct_pred + 1
        if correct_pred==len(X)+1: #no more mistakes
            break
    return w
```

Here I plot the target function and final hypothesis graph with the labelled dataset. To plot the final hypothesis graph, I have to calculate the slope of the weights and multiply it to the range. 

```python
w = PLA(X, Y)
x_ax = np.arange(-100, 100)
slope = -w[1]/w[2]
y_ax = slope*x_ax
plt.plot(x_ax, y_ax, '-',label='final hypothesis g(X)')
plt.legend()
a = np.arange(-100,100)
plt.plot(a, target(a))
for i in range(0, n):
    if y[i] == 1:
        plt.scatter(x1[i],x2[i], color = 'blue', marker = '+')
    else:
        plt.scatter(x1[i],x2[i], color = 'red', marker = '$-$')
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")
plt.title("Data set size = %s" %n)
plt.show()
```
